```python
import pandas as pd
```

```python
raw = pd.read_csv('gr-f-11.04.03-EXP-03.csv')
raw
```

```python
data = raw[['Discretion', 'Zahlen', 'Titre']].rename(columns={'Discretion': 'Year', 'Zahlen': 'Distance', 'Titre': 'Type'})
data
```

```python
plot = data.pivot(index= 'Year', columns='Type')
plot.reset_index(inplace=True)
```

```python
plot.to_csv('plot.csv')
```
